# handling missing value

```python
def missin_val(df, col):
    cate = ['Gender','Married','Dependents','Education','Self_Employed','Credit_History','Property_Area']
    if col in cate:
        df[col] = df[col].fillna(df[col].value_counts().index[0])
    else:
        df[col] = df[col].fillna(df[col].mean())
    return df
        
```

```python
df.head()
```

```python
li = df.isnull().sum()

for key in li.keys():
    if li[key] > 0:
        df = missin_val(df, key)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
cate = ['Gender','Married','Dependents','Education','Self_Employed','Credit_History','Property_Area','Loan_Status']  

for col in cate:
    df[col] = encoder.fit_transform(df[col])

df.head()
```

```python
x = df.drop(columns =['Loan_Status'])
y = df['Loan_Status']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
models = {
    'LogisticRegression': LogisticRegression(random_state=42, max_iter= 1000),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'SVC': SVC(random_state=42),
    'DecisionTreeClassifier': DecisionTreeClassifier(criterion="gini", max_depth = 1, random_state=42),
    'DecisionTreeClassifier2': DecisionTreeClassifier(criterion="entropy",max_depth = 1, random_state=42)
}
```

```python
cv =  StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state= 42)
```

```python
from sklearn.model_selection import cross_val_score

def eval_model(models, X, y):
    for name, model in models.items():
        print(name,':')
        score = np.array(cross_val_score(model,x,y, cv=cv))
        print(score.mean())
        print(score)
        print('*'*30)
        
eval_model(models, x, y)
```

we can see that logistic regression or dtc (both entorpy and gini) are good algorithm for this dataset.

we will use dtc.

```python
DTC = DecisionTreeClassifier(criterion="gini", max_depth = 1, random_state=42)
```

```python
score = np.array(cross_val_score(DTC,x,y, cv=cv))
print(score.mean())
print(score)
print('*'*30)
```
